If you've skipped the url above out of sloth, here's the game of life:
* You have a 2D grid of cells, where each cell is "alive"(1) or "dead"(0)
* Any living cell that has 2 or 3 neighbors survives, else it dies [0,1 or 4+ neighbors]
* Any cell with exactly 3 neighbors becomes alive (if it was dead)

For this task, you are given a reference numpy implementation that you must convert to pytorch.
_[numpy code inspired by: https://github.com/rougier/numpy-100]_


__Note:__ You can find convolution in `torch.nn.functional.conv2d(Z,filters)`. Note that it has a different input format. 


````python
from scipy.signal import correlate2d as conv2d

def np_update(Z):
    # Count neighbours with convolution
    filters = np.array([[1,1,1],
                        [1,0,1],
                        [1,1,1]])
    
    N = conv2d(Z,filters,mode='same')
    
    # Apply rules
    birth = (N==3) & (Z==0)
    survive = ((N==2) | (N==3)) & (Z==1)
    
    Z[:] = birth | survive
    return Z
````

````python
def torch_update(Z):
    """
    Implement an update function that does to Z exactly the same as np_update.
    :param Z: torch.FloatTensor of shape [height,width] containing 0s(dead) an 1s(alive)
    :returns: torch.FloatTensor Z after updates.
    
    You can opt to create new tensor or change Z inplace.
    """
    
    #<Your code here!>
    filters = torch.FloatTensor([[1,1,1],
                            [1,0,1],
                            [1,1,1]])

#     Z.unsqueeze_(0)
    N = F.conv2d(Z[None, None, :, :],filters[None, None, :, :], padding=1, stride=1)
    
    # Apply rules
    birth = (N==3) & (Z==0)
    survive = ((N==2) | (N==3)) & (Z==1)
    
    Z[:] = birth | survive
    
    return Z # Z[0, 0, :, :]
````

````python
#initial frame
Z_numpy = np.random.choice([0,1],p=(0.5,0.5),size=(100,100))
Z = torch.from_numpy(Z_numpy).type(torch.FloatTensor)

#your debug polygon :)
Z_new = torch_update(Z.clone())

#tests
Z_reference = np_update(Z_numpy.copy())
assert np.all(Z_new.numpy() == Z_reference), "your pytorch implementation doesn't match np_update. Look into Z and np_update(ZZ) to investigate."
print("Well done!")
````

````python
# (re-)load numpy layers
# %run ../modules/numpy_modules.ipynb
import sys
sys.path.insert(0, '../data/notMNIST_small/')
from notmnist import load_notmnist
X_train, y_train, X_test, y_test = load_notmnist(letters='ABCDEFGHIJ')
X_train, X_test = X_train.reshape([-1, 784]), X_test.reshape([-1, 784])

# Cast data to Tensors
X_train, X_test = torch.from_numpy(X_train), torch.from_numpy(X_test)
y_train, y_test = torch.from_numpy(y_train), torch.from_numpy(y_test)
````

````python
def get_batches(X, Y, batch_size=256, reshuffle=True):
    """
    Generates batches of fixed batch_size from data
    """
    assert X.size()[0] == Y.size()[0], "X and Y sample sizes mismatch!"
    n_samples = X.size()[0]
    
    #Generate new indexes
    if reshuffle:
        idx = torch.randperm(n_samples)
    else:
        idx = torch.arange(n_samples)
    
    for start in torch.range(0, n_samples, batch_size):
        
        end = torch.min(torch.IntTensor((start + batch_size, n_samples)))
        batch_idx = idx[int(start.item()):int(end.item())]
        
        yield X[batch_idx], Y[batch_idx]
    
````

````python
#Put everything together
net = nn.Sequential()

# Linear layer with 100 hidden dimencions
net.add_module('fc_1', nn.Linear(28*28, 200))
net.add_module('relu_1', nn.LeakyReLU()) 
# net.add_module('dp1', nn.Dropout())
# Output linear layer
net.add_module('fc_2', nn.Linear(200, 50))
net.add_module('relu_2', nn.LeakyReLU())

net.add_module('fc_3', nn.Linear(50, 10))
# net.add_module('lsm', nn.LogSoftmax())   # Use nn.CrossEntropyLoss instead

# Setting parameters`
lr = 3 * 1e-4 # Karpathy's constant
n_epoch = 150
batch_size = 256
opt = torch.optim.Adam(net.parameters(), lr=lr)
criterion = nn.CrossEntropyLoss()
````

### Data preproccessing

````python
# Create OHE for y_train and y_test
y_train_ohe = torch.zeros(len(y_train), y_train.max()+1).scatter_(1, y_train.unsqueeze(1), 1.)
y_test_ohe = torch.zeros(len(y_test), y_test.max()+1).scatter_(1, y_test.unsqueeze(1), 1.)
````

````python
loss_history = []
test_loss_history = []
for i in range(n_epoch):
    for x_batch, y_batch in get_batches(torch.tensor(X_train, dtype=torch.float32),\
                                        y_train, batch_size):
          
        # Forward
        y_scores = net(x_batch)
        loss = criterion(y_scores, y_batch)
        
#         loss_history.append(loss)
    
        # Compute gradients
        loss.backward()
        
        # Adam step
        opt.step()
        
        #clear gradients
        opt.zero_grad()

    loss_history.append(loss)
    
    y_scores_test = net(X_test)
    loss_test = criterion(y_scores_test, y_test)
    test_loss_history.append(loss_test)
    # Visualize
    display.clear_output(wait=True)
    plt.figure(figsize=(8, 6))
        
    plt.title("Training and Test loss")
    plt.xlabel("#iteration")
    plt.ylabel("loss")
    plt.plot(loss_history, 'b')
    plt.plot(test_loss_history, 'g')
#     plt.yscale('log')
    plt.show()
    
    print('Current train loss: %f \n Current test loss: %f' % (loss, loss_test)  
````

Let's see how we perform on test data
