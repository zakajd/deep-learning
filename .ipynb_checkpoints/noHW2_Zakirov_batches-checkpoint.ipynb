```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

```python
# класс one-hot encoder'а для MNIST
class OneHotEncoder:
    def __init__(self):
        self.transform_mapping = np.zeros((10,10))
        for i in range(self.transform_mapping.shape[0]):
            self.transform_mapping[i][i] = 1.0
    def transform(self, y):
        return self.transform_mapping[int(y)]
```

```python
def make_submission(X_test, net, fname="my_submission.csv"):
    with open(fname,'w') as fout:
        fout.write('Id,Category')
        for i in range(X_test.shape[0]):
            y_h = net.forward(X_test[i])
            y = np.argmax(y_h)
            fout.write("\n{},{}".format(i, int(y)))
```

```python
# внимание, тут Y_test в обычном формате(не onehot)
def compute_acc(X_test, Y_test, net):
    acc = 0.0
    for i in range(X_test.shape[0]):
        y_h = net.forward(X_test[i])
        y = np.argmax(y_h)
        if(y == Y_test[i]):
            acc += 1.0
    return acc / Y_test.shape[0]
```

```python
# Определение слоев сети
class Dense(object):
    def __init__(self, in_size, out_size, rlambda=0.001): #

        self.weights = np.random.normal(scale=1, size=(out_size, in_size)) * np.sqrt(2 / in_size)
        print(self.weights.dtype)
        #self.bias = np.random.randint(0, 1, size=out_size,)
        self.bias = np.zeros(out_size)
        self.rlambda = rlambda # regularization parameter
    
    def get_reg_loss(self):
        return 0.5 * self.rlambda * (np.linalg.norm(self.W, ord='fro') ** 2)
  
    def forward(self, X):
        self.X = X
        # X should be a column vector or array with columns containing images.
        return np.dot(self.weights, X) + self.bias
    
    def backward(self, dz, lr=0.001):
        """Update weights and biases using backpropagation"""
        
        # self.dw = np.outer(dz, self.X) # ???
        self.dw = np.matmul(dz, self.X.T)
        print(self.dw.shape)
        self.db = dz #* np.ones(self.bias.shape)
        
#         self.dX = np.matmul(dz, self.weights)
        self.dX = np.matmul(self.weights.T, dz)
        if self.rlambda != 0:
            self.dw += self.rlambda * self.weights
        
        #Update weights
        print(self.weights.dtype, self.dw.dtype)
        print(self.weights.shape, self.dw.shape)
        self.weights -= (lr / len(dz)) * self.dw
        print(self.bias.dtype, self.db.dtype)
        print(self.bias.shape, self.db.shape)
        self.bias -= (lr / len(dz)) * np.sum(self.db, axis=0)
        return self.dX
    
class Softmax(object): 
    """Vanilla Softmax layer for Neural Network"""
    
    def forward(self, X):
        """"a^(L) activitation vector from L layer 
        C used to prevent exponents explosion """
        # Find max for each column
        C = - np.max(X, axis=0)
        self.exps = np.exp(X + C)
        
        #sum of elements in each column == 1
        self.sf = self.exps / np.sum(self.exps, axis=0)
        return self.sf
        
    
    def backward(self, dz):
        length = len(dz)
        dx = np.zeros(dz.shape)
        
        #iterate columns containg softmax values for each individual example
        for i in range(len(self.sf.T)):
            k = self.sf.T[i]
            # k is a raw version of column vector
            #matrix with rows 
            matrix = np.matmul(k.reshape(-1,1), np.ones((1, length))) * \
                    (np.identity(length) - \
                         np.matmul(np.ones((length, 1)), k.reshape(1,-1)))
            dx[:, i] = np.sum(matrix, axis=1) #sum by rows                  
    
        return dx * dz 
        

class CrossEntropy(object):

    def forward(self, y_true, y_hat):
        self.y_true = y_true
        self.y_hat = y_hat
        self.ziped = zip(y_true, y_hat)
        #Sum over hole matrix to get total loss for 
        return - np.sum([x[0]*np.log(x[1]) for x in self.ziped])

    def backward(self, dz): #, lr=0.001):
        return (- self.y_true / self.y_hat) * dz

class ReLU(object):
    def forward(self, X):
        self.X = X    
        #return max(0, X)
        return np.maximum(0, X)
    
    def backward(self, dz):
        dz[self.X < 0] = 0
        return dz
```

```python
# Опеределение самой сети
# class MnistNet
class MnistNet(object):
    def __init__(self):
        self.d1 = Dense(784,16) # rlambda?
        self.a1 = ReLU()
#         self.d2 = Dense(16,16)
#         self.a2 = ReLU()
        self.d3 = Dense(16,10)
        self.sm = Softmax()
#         self.ce = CrossEntropy()
    
    def forward(self, X):
        # Transpose X to get column vectors of examples
        net = self.d1.forward(X)
        a = self.a1.forward(net)
#         net = self.d2.forward(a)
#         a = self.a2.forward(net)
        net = self.d3.forward(a)
        net = self.sm.forward(net)
#         net = self.ce.forward(net)
        self.net = net
        return net
    
    def backward(self, dz, lr=0.01):
#         dz = self.ce.backward(dz)
        dz = self.sm.backward(dz)
        dz = self.d3.backward(dz, lr)
        dz = self.a1.backward(dz, lr)
        dz = self.d1.backward(dz, lr)
        return dz
```

```python
# Здесь процедура обучения
# X_train, X_test, Y_train, Y_test
net = MnistNet()
loss = CrossEntropy()
lr = 0.001
epochs = 3
mini_batch_size = 256
L_train = []
L_test = []

for iter in range(epochs):
    L_acc = 0.
    L_test_acc = 0.
    # Shuffle data and split into mini-batches
    sh_ind = list(range(X_train.shape[0]))
    np.random.shuffle(sh_ind)
    mini_batches = [sh_ind[i:i+mini_batch_size] \
                    for i in range(0, X_train.shape[0], mini_batch_size)]
#     mb_data = [X_train[data] for data in mini_batches]
#     mb_labels = [Y_train[ind] for ind in mini_batches]
    
    for mini_batch in mini_batches:
        X = X_train[mini_batch]
        Y = Y_train[mini_batch]
        #Count loss inside Neural network?
        Y_hat = net.forward(X.T)
        print(type(Y_hat))
        L = loss.forward(Y.T, Y_hat) #+ net.get_reg_loss()
        L_acc += L # sum loss for all training exaples
        dz = loss.backward(1)
        dz = net.backward(dz, lr)
    
    L_acc /= Y_train.shape[0] # X_train.shape[0] ?
    L_train.append((iter, L_acc))
    
    for i in range(X_test.shape[0]):
        x = X_test[i]
        y = Y_test[i]
        y_h = net.forward(x.T)
        L = loss.forward(y.T, y_h) #+ net.get_reg_loss()
        L_test_acc += L
    L_test_acc /= Y_test.shape[0]
    L_test.append((iter, L_test_acc))
    
    print(f"{iter} epoch loss. Train : {L_acc} . Test : {L_test_acc}")
    

```
